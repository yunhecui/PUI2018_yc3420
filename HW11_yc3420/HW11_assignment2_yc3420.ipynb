```python
import geopandas as gpd
import urllib
import os
import pylab as pl
import zipfile

%pylab inline
```

```python
#download the data1
urllib.request.urlretrieve('https://data.cityofnewyork.us/download/i8iw-xf4u/application%2Fzip', 'ZIPCODE.gz')
```

```python
zipref = zipfile.ZipFile(os.getenv("PUIDATA") + "/ZIPCODE.gz", 'r')
zipref.extractall(os.getenv("PUIDATA") + '/ZIP')
zipref.close()
```

```python
path = os.getenv("PUIDATA") + '/ZIP'
namelist = os.listdir(path)
print (namelist)
```

```python
def getshp(path):
    
    nl = os.listdir(path)
    for n in nl:
        if os.path.splitext(n)[-1] == '.shp':
            shp_nm = str(os.path.basename(n))
    return shp_nm
            
                    
path2 = os.getenv("PUIDATA") + '/ZIP'
getshp(path2)
```

```python
path_shp = os.getenv("PUIDATA") + '/ZIP/' + getshp(path)
```

```python
zipshp = gpd.GeoDataFrame.from_file(path_shp)
```

```python
zipshp.head()
```

```python
zipshp['Pop_sqml'] = zipshp['POPULATION'] / (zipshp['AREA'] * 3.587e-8)
```

```python
zipshp.shape
```

```python
zipshp = zipshp[zipshp['Pop_sqml'] < 200000]
```

```python
zipshp['Pop_sqml'].describe()
```

```python
zipshp.head()
```

```python
zipshp_mer = zipshp.to_crs(epsg = 4326)
m1 = zipshp_mer.plot(column = 'Pop_sqml', figsize = (12,12), cmap = 'viridis', legend = 'True')
m1.set_title("NYC population density by zip, per square mile")
m1.set_xlabel("longitude")
m1.set_ylabel("latitude")
```

Fig.1 This map shows the New York City population density by zip. The darker the color is, the larger the population density is.  The population density is in square mile.
